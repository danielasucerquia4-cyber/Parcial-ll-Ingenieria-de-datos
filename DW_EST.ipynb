# 📚 Construcción de un Data Warehouse Simple en Python (con Pandas)

En este tutorial, aprenderás a integrar múltiples tablas relacionadas en un único **Data Warehouse** usando el modelo estrella (*star schema*). Este es un patrón común en inteligencia de negocios, donde una **tabla de hechos** (ventas) se enriquece con varias **tablas de dimensiones** (clientes, productos, fechas).

### Paso 1: Importar las librerías necesarias

Lo primero es traer las herramientas que usaremos. Para este caso, solo necesitamos `pandas`, la librería estándar para manipulación de datos en Python.

```python
# Importamos pandas, que nos permite trabajar con DataFrames (tablas de datos en memoria).
import pandas as pd
```

### Paso 2: Cargar los datasets desde archivos CSV

Tenemos cuatro archivos de datos separados. Cada uno representa una entidad del modelo de datos.

```python
import pandas as pd

# Cargamos la tabla de hechos: 'ventas'.
# Contiene los eventos transaccionales (las ventas realizadas).
ventas = pd.read_csv('ventas.csv')

# Cargamos las tablas de dimensiones.
# Cada una describe un aspecto de las ventas.
clientes = pd.read_csv('clientes.csv') # Información sobre quién compró.
productos = pd.read_csv('productos.csv') # Información sobre qué se compró.
fechas = pd.read_csv('fechas.csv') # Información sobre cuándo se compró.
```

### Resumen del flujo de trabajo

1.  **Cargar** los datos desde sus fuentes.
